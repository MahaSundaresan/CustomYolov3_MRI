```python
import os, sys
import random
from scipy import ndarray

import skimage as sk
from skimage import transform
from skimage import util
from skimage import io
```

```python
# Data augmentation of images

# types of transformation for augmentation

def rotation_45(image_array: ndarray):
    return sk.transform.rotate(image_array, 45)


def rotation_90(image_array: ndarray):
    return sk.transform.rotate(image_array, 90)


def rotation_180(image_array: ndarray):
    return sk.transform.rotate(image_array, 180)



def gaussian_noise(image_array: ndarray):
     return sk.util.random_noise(image_array, mode='gaussian', seed=None, clip=True)

# listing the transformations used
augment_by_transform = {
    'rotate_45': rotation_45,
    'rotate_90': rotation_90,
    'rotate_180': rotation_180,
    'gaussian_noise': gaussian_noise,
}

folder_path = r"Walkingperson"
#number of augmented images needed
augmented_images_count = 25

# extracting image files from the folder
# Reference : https://stackoverflow.com/questions/973473/getting-a-list-of-all-subdirectories-in-the-current-directory/32505092
images = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
augmented_image_gen= 1

while augmented_image_gen <= augmented_images_count:
    # choosing a random image from the folder
    img_path = random.choice(images)
    img_transform = sk.io.imread(img_path)
    
    transformed_image = None
    # choosing a random transformation from the list to be applied on the image
    key = random.choice(list(augment_by_transform))
    after_transform_image = augment_by_transform[key](img_transform)
    #converting to uint8 prior to saving to suppress warning due to lossy conversion
    t_image= sk.util.img_as_ubyte(after_transform_image, force_copy=False)

    out_path = '%s/augmented_image_%s%s.jpg' % (folder_path, augmented_image_gen, key)

    io.imsave(out_path, t_image)
    augmented_image_gen += 1
```
